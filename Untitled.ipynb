```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
data = pd.read_csv('data_dt_ds_2023.csv', sep=';', encoding='utf-8')

# 1. Số lượng tỉnh thành được thống kê trong file data
so_luong_tinh_thanh = len(data)
print(f"Số lượng tỉnh thành được thống kê trong file data: {so_luong_tinh_thanh}")
```

```python
# 2. Liệt kê 5 tỉnh thành có diện tích nhỏ nhất và lớn nhất
top_5_dien_tich_nho_nhat = data.sort_values(by=' Diện tích (km vuông) ', ascending=True).head(5)
top_5_dien_tich_lon_nhat = data.sort_values(by=' Diện tích (km vuông) ', ascending=False).head(5)
print("\n5 tỉnh thành có diện tích nhỏ nhất:")
print(top_5_dien_tich_nho_nhat[[' Thành phố/Tỉnh ', ' Diện tích (km vuông) ']])
print("\n5 tỉnh thành có diện tích lớn nhất:")
print(top_5_dien_tich_lon_nhat[[' Thành phố/Tỉnh ', ' Diện tích (km vuông) ']])
```

```python
# 3. Liệt kê 5 tỉnh thành dân số nhỏ nhất và lớn nhất
top_5_dan_so_nho_nhat = data.sort_values(by=' Dân số (người)', ascending=True).head(5)
top_5_dan_so_lon_nhat = data.sort_values(by=' Dân số (người)', ascending=False).head(5)
print("\n5 tỉnh thành có dân số nhỏ nhất:")
print(top_5_dan_so_nho_nhat[[' Thành phố/Tỉnh ', ' Dân số (người)']])
print("\n5 tỉnh thành có dân số lớn nhất:")
print(top_5_dan_so_lon_nhat[[' Thành phố/Tỉnh ', ' Dân số (người)']])
```

```python
# 4. Vẽ biểu đồ đường của 10 tỉnh thành có dân số lớn nhất
top_10_dan_so_lon_nhat = data.sort_values(by=' Dân số (người)', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.plot(top_10_dan_so_lon_nhat[' Thành phố/Tỉnh '], top_10_dan_so_lon_nhat[' Dân số (người)'])
plt.xlabel('Tỉnh thành')
plt.ylabel('Dân số (người)')
plt.title('Biểu đồ đường dân số của 10 tỉnh thành có dân số lớn nhất')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5. Vẽ biểu đồ tròn phần trăm thể hiện tỷ trọng diện tích của 5 tỉnh thành có diện tích lớn nhất
top_5_dien_tich_lon_nhat = data.sort_values(by=' Diện tích (km vuông) ', ascending=False).head(5)
tong_dien_tich = data[' Diện tích (km vuông) '].sum()
labels = [
    f"{top_5_dien_tich_lon_nhat.iloc[i][' Thành phố/Tỉnh ']}" 
    for i in range(len(top_5_dien_tich_lon_nhat))
] + ['Các tỉnh thành còn lại']
sizes = [
    top_5_dien_tich_lon_nhat.iloc[i][' Diện tích (km vuông) '] 
    for i in range(len(top_5_dien_tich_lon_nhat))
] + [tong_dien_tich - sum(top_5_dien_tich_lon_nhat[' Diện tích (km vuông) '])]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Tỷ trọng diện tích của 5 tỉnh thành có diện tích lớn nhất')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# 6. Hiện thông tin thứ hạng bao nhiêu về diện tích và dân số so với các tỉnh còn lại của một tỉnh thành bất kỳ được nhập từ bàn phím
tinh_thanh = input("Nhập tên tỉnh thành bạn muốn tìm: ")
try:
    # Tìm thông tin của tỉnh thành đó trong data
    row = data[data[' Thành phố/Tỉnh '] == tinh_thanh]
    if row.empty:
        print(f"Không tìm thấy thông tin tỉnh thành '{tinh_thanh}' trong dữ liệu.")
    else:
        # Xác định thứ hạng về diện tích
        thu_hang_dien_tich = data[' Diện tích (km vuông) '].rank(ascending=False, method='dense').loc[data[' Thành phố/Tỉnh '] == tinh_thanh].values[0]
        # Xác định thứ hạng về dân số
        thu_hang_dan_so = data[' Dân số (người)'].rank(ascending=False, method='dense').loc[data[' Thành phố/Tỉnh '] == tinh_thanh].values[0]
        print(f"Tỉnh thành '{tinh_thanh}' có thứ hạng:")
        print(f"- Thứ hạng diện tích: {int(thu_hang_dien_tich)}/63")
        print(f"- Thứ hạng dân số: {int(thu_hang_dan_so)}/63")
except Exception as e:
    print(f"Lỗi: {e}")
```
